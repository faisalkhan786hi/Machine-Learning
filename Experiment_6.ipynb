```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics

# Step 1: Load the dataset
df = pd.read_csv('/content/Engine expt6.csv')

# Step 2: Define the independent variables (features) and dependent variable (target)
X = df[['ENGINESIZE', 'CYLINDERS', 'FUELCONSUMPTION_CITY', 'FUELCONSUMPTION_HW']]
y = df['CO2EMISSIONS']

# Step 3: Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Step 4: Train the linear regression model on the training data
regressor = LinearRegression()
regressor.fit(X_train, y_train)

# Step 5: Predict the target variable using the trained model on the testing data
y_pred = regressor.predict(X_test)

# Step 6: Evaluate the performance of the model using metrics such as mean squared error (MSE) and coefficient of determination (R-squared)
print('Answer 2 >>>>  Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('R-squared:', metrics.r2_score(y_test, y_pred))
```

Answer 3 :

In this model, I have used the following four attributes as the independent variables (features):

ENGINESIZE
CYLINDERS
FUELCONSUMPTION_CITY
FUELCONSUMPTION_HW

These attributes represent the engine size, number of cylinders, fuel consumption in the city, and fuel consumption on the highway, respectively. These attributes are commonly used as predictors of CO2 emissions in vehicles, as they are known to have a strong correlation with the amount of CO2 emitted by the vehicle.
