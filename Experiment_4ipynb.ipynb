# Predict CO2 Emission using Simple Linear **Regression**

1. Scatter plot “EngineSize vs CO2Emissions”


```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/content/CO2 Emission.csv')
plt.scatter(df["EngineSize"], df["C02Emissions"])
plt.xlabel("Engine Size")
plt.ylabel("CO2 Emissions")
plt.title("Engine Size vs CO2 Emissions")
plt.show()
```

2. Predict the value of CO2 emission on the basis of Engine size for enginesize=2.4. using data 
given in “CO2Small.csv” file.


```python
from sklearn.linear_model import LinearRegression

# Fit a linear regression model to the data
model = LinearRegression()
model.fit(df[["EngineSize"]], df["C02Emissions"])

# Use the model to predict CO2 emissions for an engine size of 2.4
engine_size = 2.4
predicted_emissions = model.predict([[engine_size]])
print("Predicted CO2 emissions for engine size of {}: {:.2f}".format(engine_size, predicted_emissions[0]))
```

3. Calculate R^2 .


```python
r_squared = model.score(df[["EngineSize"]], df["C02Emissions"])
print("R^2 value: {:.2f}".format(r_squared))
```

4. Calculate the above details on the larger dataset i.e. file “CO2Full Data.csv”


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load the data from the CSV file into a pandas DataFrame
data = pd.read_csv("/content/CO2.csv")

# Plot a scatter plot of EngineSize vs CO2Emissions
plt.scatter(data["ENGINESIZE"], data["CO2EMISSIONS"])
plt.xlabel("Engine Size")
plt.ylabel("CO2 Emissions")
plt.title("Engine Size vs CO2 Emissions")
plt.show()

# Perform linear regression to find the best-fit line
x = data[["ENGINESIZE"]]
y = data["CO2EMISSIONS"]
model = LinearRegression().fit(x, y)
r_sq = model.score(x, y)
y_pred = model.predict(x)

# Print the coefficients and R^2 value of the regression line
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
print("R^2:", r_sq)

# Plot the best-fit line along with the data points
plt.scatter(x, y)
plt.plot(x, y_pred, color="red")
plt.xlabel("Engine Size")
plt.ylabel("CO2 Emissions")
plt.title("Engine Size vs CO2 Emissions")
plt.show()
```

5. Print the final equation of the line


```python
x = data[["ENGINESIZE"]]
y = data["CO2EMISSIONS"]
model = LinearRegression().fit(x, y)
# Print the coefficients and intercept of the regression line
print("Equation of the line: CO2 Emissions = {:.2f} * Engine Size + {:.2f}".format(model.coef_[0], model.intercept_))
```

6. Plot final line (best fit line) along with other data points.


```python
# Perform linear regression to find the best-fit line
x = data[["ENGINESIZE"]]
y = data["CO2EMISSIONS"]
model = LinearRegression().fit(x, y)
r_sq = model.score(x, y)
y_pred = model.predict(x)

# Plot the data points and the best-fit line
plt.scatter(x, y)
plt.plot(x, y_pred, color="red")
plt.xlabel("Engine Size")
plt.ylabel("CO2 Emissions")
plt.title("Engine Size vs CO2 Emissions (Full Dataset)")
plt.show()
```
