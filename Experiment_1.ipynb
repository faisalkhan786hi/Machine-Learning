[a.]    Load this file into a python DataFrame

```python
import pandas as pd

df = pd.read_csv("/content/student.csv")
print(df)
```

[b.] Find the number of rows and columns in it

```python
df.shape
```

[c.] Print column names


```python
df.columns
```

[d.] Change column name 'Name' with new name 'FirstName'


```python
df.rename(columns={'Name': 'FirstName'})
```

[e.] Print last 5 rows from the bottom


```python
df.tail(5)
```

[f.] Print the details of student with lowest marks.


```python
min_marks_students = df.query('Marks == Marks.min()').nsmallest(n=len(df), columns=['Marks'])
print(min_marks_students)
```

[g.] Find total marks of all female students


```python
total_marks_female = df.loc[df['Sex'] == 'F', 'Marks'].sum()
print(total_marks_female)
```

[h.] List names of all the male students


```python
male_students = df.loc[df['Sex'] == 'M', 'Name'].tolist()
print(male_students)
```

[i.] Find mean age of the class


```python
mean_age = df['Age '].mean()

print("Mean age of the class:", mean_age)
```

[j.] Line plot marks of the class


```python
import matplotlib.pyplot as plt
plt.plot(df['Marks'])
plt.xlabel('Student')
plt.ylabel('Marks')
plt.title('Marks of the Class')
plt.show()
```

[k.] Find the index of record of oldest student in the class


```python
oldest_student_index = df['Age '].idxmax()

print("Index of the oldest student:", oldest_student_index)
```

[l.] sort and print the data on the basis of Name followed by Age.


```python
sorted_df = df.sort_values(['Name', 'Age '])
print(sorted_df)
```

[m.] Change the name 'Nihar' to 'Jason Bourne' in name column of the DataFrame.


```python
df.loc[df['Name'] == 'Nihar', 'Name'] = 'Jason Bourne'
print(df)
```

[n.] Change and print order of the columns (Name, Sex, Age, Marks, Grade)


```python
df = df[['Name', 'Sex', 'Age ', 'Marks', 'Grade']]
print(df)
```

[o.] Count and print number of students sex wise and display result with suitable column 
headers


```python
sex_count = df.groupby('Sex').size().reset_index(name='Count')
print(sex_count)
```

[p.] Delete and print row where age=46


```python
df = df[df['Age '] != 46]
print(df)
```

[q.] Print the data types of individual columns of the data frame


```python
print(df.dtypes)
```

[r.] Convert and print the datatype of a given column Age (int to float).


```python
df['Age '] = df['Age '].astype(float)
print(df.dtypes)
```

[s.] Create a new column named “UpdatedMarks” which as 5.5% more marks than the 
existing Marks column.


```python
df['UpdatedMarks'] = df['Marks'] * 1.055
print(df)
```

[t.] Delete the “Marks Column”

```python
df = df.drop(columns=['Marks'])
print(df)
```
