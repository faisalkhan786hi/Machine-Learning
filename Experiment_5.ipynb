# Predict CO2 Emission using Gradient Descent.

Screenshot 2023-04-04 020855.png

BUILDING GRADIENT DESCENT MODEL

```python
# Building the model
import numpy as np
import pandas as pd

# Load the data and extract the input features and target variable
df = pd.read_csv('/content/CO2 Emission.csv')
m = 0
c = 0

L = 0.0001  # The learning Rate
epochs = 1000  # The number of iterations to perform gradient descent

x = df['EngineSize'].values
y = df['C02Emissions'].values

# Performing Gradient Descent 
for i in range(epochs): 
    Y_pred = m*x + c  # The current predicted value of Y
    D_m = (-2/9) * sum(x * (y - Y_pred))  # Derivative wrt m
    D_c = (-2/9) * sum(y - Y_pred)  # Derivative wrt c
    m = m - L * D_m  # Update m
    c = c - L * D_c  # Update c
    
print (m, c)
```

1. Scatter plot “EngineSize vs CO2Emissions”


```python
import matplotlib.pyplot as plt

Y_pred = m*x + c
plt.scatter(x, y)
plt.plot([min(x), max(x)], [min(Y_pred), max(Y_pred)], color='red') # predicted
plt.show()
```

2. Calculate R^2 .


```python
from sklearn.metrics import r2_score
from sklearn import metrics as mt
from sklearn.metrics import mean_squared_error
score=r2_score(y,Y_pred)
rms = np.sqrt(mean_squared_error(y,Y_pred))
print(">>>>R2 score is:",score)
print("RMS value is:",rms)
print("The Mean Absolute Error of model is:", np.round(mt.mean_absolute_error(y,Y_pred),2))
```

3. Print the final equation of the line


```python
# Printing the final equation of the line
print(f"The final equation of the line is: C02Emissions = {m:.4f} * EngineSize + {c:.4f}")
```

4. Plot final line (best fit line) along with other data points.

```python
import matplotlib.pyplot as plt

# Plotting the data points
plt.scatter(x, y)

# Plotting the final line (best fit line)
plt.plot(x, m*x+c, color='red')

# Adding labels and title
plt.xlabel('Engine Size')
plt.ylabel('CO2 Emissions')
plt.title('Linear Regression Model')

# Displaying the plot
plt.show()
```
